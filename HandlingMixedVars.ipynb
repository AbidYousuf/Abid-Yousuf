```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.sample(4)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers travelling with')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df["number"],errors='coerce',downcast='integer')
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['C_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['C_cat'] = df['Cabin'].str[0] # captures the first letter
```

```python
df.head()
```

```python
df['C_cat'].value_counts().plot(kind='bar')
```

```python
df.sample(3)
```

```python
df['T_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['T_num'] = pd.to_numeric(df['T_num'],
                                   errors='coerce',
                                   downcast='integer')
df['T_num'].head(3)
```

```python
# extract the last bit of ticket as number
df['T_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['T_num'] = pd.to_numeric(df['T_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['T_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['T_cat'] = np.where(df['T_cat'].str.isdigit(), np.nan,
                              df['T_cat'])

df.head(20)
```

# **Handling Date and Time**

```python
date = pd.read_csv('/content/orders.csv')
time = pd.read_csv('/content/messages.csv')
```

**Working with Dates**

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

**1. Extract year**

```python
date['date_year'] = date['date'].dt.year
```

```python
date.sample(4)
```

**2. Extract Month**

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_month'] = date['date'].dt.month_name()
date.head()
```

**3.Extract Days**

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['date_day'] = date['date'].dt.day_name()
date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

**4.Extract week of the year**

```python
# date['date_week'] = date['date'].dt.week
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

**5.Extract Quarter**

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

**6.Extract Semester**

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

**7.Extract Time elapsed between dates**

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
