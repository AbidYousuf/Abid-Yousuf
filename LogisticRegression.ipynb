# **Perceptron Trick**

```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=50)
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  l_rate = 0.01
  for i in range(1000):
    rd = np.random.randint(X.shape[0])
    y_hat = step(np.dot(X[rd],weights))
    weights = weights + l_rate*(y[rd]-y_hat)*X[rd]
  return weights[0],weights[1:]
```

```python
np.dot(np.insert(X[35],0,1),[1,1,1])
```

```python
def step(z):
  if z>0:
    return 1
  else:
    return 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(intercept_,' ',coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = - (intercept_/coef_[1])
```

```python
x_input = np.linspace(-5,5,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(6,5))
plt.plot(x_input,y_input,color='red',linewidth=2)
plt.scatter(X[:,0],X[:,1],c=y,cmap='plasma',s=50)
plt.ylim(-3,2)
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  l_rate = 0.01
  for i in range(1000):
    rd = np.random.randint(X.shape[0])
    y_hat = sigmoid(np.dot(X[rd],weights))
    weights = weights + l_rate*(y[rd]-y_hat)*X[rd]
  return weights[0],weights[1:]
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(intercept_,' ',coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m*x_input2 + b
```

# **Logistic Regression using Gradient Descent**

```python
def gradient_descent(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  l_rate=0.01
  for i in range(900000):
    y_hat = sigmoid(np.dot(X,weights))
    weights = weights + l_rate*(np.dot(y-y_hat,X))/X.shape[0]
  return weights[0],weights[1:]
```

```python
intercept_,coef_ = gradient_descent(X,y)
```

```python
print(intercept_,' ',coef_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input3 = np.linspace(-3,3,100)
y_input3 = m*x_input3 + b
```
